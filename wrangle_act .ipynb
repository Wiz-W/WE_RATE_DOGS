## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

```python
# Make copies of original pieces of data
twitter_archive_clean = twitter_archive.copy()
image_predictions_clean = image_predictions.copy()
data_clean = data.copy()
```

### Issue #1:

#### Define
    Convert Timestamp from object to DateTime dtype in twitter_archive table.

#### Code:

```python
# conversion of timestamp object to datetime dtype
twitter_archive_clean['timestamp'] = pd.to_datetime(twitter_archive_clean['timestamp'])
```

### Issue #2:

#### Define
    Slicing the html Tag present in the source column of the Twitter_archive table. 

#### Code

```python
#creating a list of iterms will would like to remove
tags_list = ['<p>' ,'</p>' , '<p*>',
             '<ul>','</ul>',
             '<li>','</li>',
             '<br>',
             '<span*>','</span>',
             '<a href*>','</a>',
             '<em>','</em>','<a']

for tag in tags_list:
    twitter_archive_clean['source'].replace(to_replace=tag, value='', regex=True, inplace=True)
```

```python
twitter_archive_clean['source'][0]
```

## Analyzing and Visualizing Data
In this section, analyze and visualize your wrangled data. You must produce at least **three (3) insights and one (1) visualization.**

```python
twitter_archive_master = pd.read_csv(r'C:\Users\User\Desktop\Data analytics s\UDACITY\UDACITY PROJECTS\project2\twitter_archive_master.csv')
import matplotlib.pyplot as plt
%matplotlib inline
```

### Insights:
1.Top 10 Dog breeds with the highest number of Retweet counts

2.Top 10 dog breeds with the highest rating

3.Top 10 dog breeds with the highest favorite_count

### Analysis 1:Top 10 Dog breeds with the highest number of Retweet counts

```python
top_10_dogBreed_by_retweetCount = twitter_archive_master.sort_values(by = ['retweet_count'],ascending = False)[['animal_breed1','retweet_count']].head(10)
top_10_dogBreed_by_retweetCount.reset_index(drop=True,inplace=True)
top_10_dogBreed_by_retweetCount.set_index('animal_breed1',inplace=True)
top_10_dogBreed_by_retweetCount
```

### Visual 1

```python
#plotting a bar graph of the  top 10 animal_breed by retweet_count 
top_10_dogBreed_by_retweetCount.plot(kind='barh')
plt.title('TOP 10  ANIMAL_BREED BY RETWEET_COUNT')
plt.ylabel('ANIMAL_BREED')
plt.xlabel('RETWEET_COUNT');
```

From the above analysis,it is shown/calculated that "Labrador_retriever" which is a dog breed has the highest retweet_count of 79515, and "golden_retriever" has the lowest retweet_count of 27680.

### Analysis 2:Top 10 dog breeds with the highest rating

```python
top_10_dogBreed_by_highestRating = twitter_archive_master.sort_values(by = ['dogs_ratings'],ascending = False)[['animal_breed1','dogs_ratings']].head(10)
top_10_dogBreed_by_highestRating.reset_index(drop=True,inplace=True)
top_10_dogBreed_by_highestRating.set_index('animal_breed1',inplace=True)
top_10_dogBreed_by_highestRating
```

#### Visual 2:

```python
#plotting a bar graph of the  top 10 animal_breed by rating 
top_10_dogBreed_by_highestRating.plot(kind='barh')
plt.title('TOP 10 ANIMAL_BREED BY RATINGS')
plt.ylabel('ANIMAL_BREED')
plt.xlabel('RATINGS');
```

From the above analysis it is shown that "Labrador_retriever" which is describe as a dog_breed happen to be the breed with the highest number of ratings with a total of 165,and "Siberian_husky" the lowest rating of 50.

#### Analysis 3:Top 10 dog breeds with the highest favorite_count

```python
top_10_dogBreed_by_highest_favorite_count = twitter_archive_master.sort_values(by = ['favorite_count'],ascending = False)[['animal_breed1','favorite_count']].head(10)
top_10_dogBreed_by_highest_favorite_count.reset_index(drop=True,inplace=True)
top_10_dogBreed_by_highest_favorite_count.set_index('animal_breed1',inplace=True)
top_10_dogBreed_by_highest_favorite_count
```

#### Visual 3:

```python
#plotting a bar graph of the  top 10 animal_breed by favorite_count
top_10_dogBreed_by_highest_favorite_count.plot(kind='barh')
plt.title('TOP 10 ANIMAL_BREED BY FAVORITE_COUNT')
plt.ylabel('ANIMAL_BREED')
plt.xlabel('FAVORITE_COUNT');
```

From the above analysis it is shown that "Lakeland_terrier" which is describe as a dog_breed happen to be the animal_breed with the highest number of favorite_count with a total of 132810,and "Chesapeake_Bay_retriever" the lowest rating of 71144.
